# Visualization of topics with wordclouds

```python
%%bash

for i in `seq 10 1 20`
do
    input=`echo "../topics_mallet/crun1/speeches_"$i"_wordsw_top50.dat"`
    dir=`echo "../topics_mallet/topic_no_"$i`
    cat $input | awk '{printf("%s:%.4f\n",$2,$3) > "file_"$1}'
    mv file* $dir
done
```

```python
%%bash

for dir in `ls -d ../topics_mallet/topic_no_*`
do
    echo $dir
    for file in `ls $dir`
    do
        ./WordCloudCairo.exe -i $dir"/"$file -o $dir"/"$file".png" -c reddish.xml
    done

done
```
